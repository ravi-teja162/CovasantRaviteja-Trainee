```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling settings
Fs = 1000          # Sampling frequency (samples per second)
T = 1/Fs           # Sampling period (time between samples)
t = np.arange(0, 1, T)  # Time vector for 1 second

# Create sine waves
f1 = 5    # Frequency 1 (Hz)
f2 = 50   # Frequency 2 (Hz)
signal = np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

# Plot the time-domain signal
plt.figure(figsize=(12, 4))
plt.plot(t, signal)
plt.title("Time Domain Signal (5Hz + 50Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()

# Perform FFT
n = len(signal)
fft_vals = np.fft.fft(signal)
fft_freqs = np.fft.fftfreq(n, T)

# Take only positive frequencies
fft_vals = np.abs(fft_vals[:n // 2])
fft_freqs = fft_freqs[:n // 2]

# Plot the frequency-domain (FFT)
plt.figure(figsize=(12, 4))
plt.stem(fft_freqs, fft_vals)
plt.title("Frequency Domain (FFT of Signal)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
#D.Create two frequencies 5Hz and 50Hz sin signals

import numpy as np
import matplotlib.pyplot as plt

samples = 1000
time = 1/samples
time_vector = np.arange(0,1,time)

f1 = 5
f2 = 50

signal_5hz = np.sin(2 * np.pi * f1 * time_vector)
signal_50hz = np.sin(2 * np.pi * f2 * time_vector)

plt.figure(figsize=(12, 5))

plt.subplot(2, 1, 1)
plt.plot(time_vector, signal_5hz)
plt.title("5 Hz Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.xlim(0, 1)

plt.subplot(2, 1, 2)
plt.plot(time_vector, signal_50hz)
plt.title("50 Hz Sine Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.xlim(0, 1)

plt.tight_layout()
plt.show()

n = len(signal_5hz)
fft_5hz = np.fft.fft(signal_5hz)
freqs_5hz = np.fft.fftfreq(n, time)
fft_5hz = np.abs(fft_5hz[:n // 2])
freqs_5hz = freqs_5hz[:n // 2]
fft_50hz = np.fft.fft(signal_50hz)
freqs_50hz = np.fft.fftfreq(n, time)
fft_50hz = np.abs(fft_50hz[:n // 2])
freqs_50hz = freqs_50hz[:n // 2]

#FFTs 

plt.subplot(2, 1, 1)
plt.stem(freqs_5hz, fft_5hz, basefmt=" ")
plt.title("FFT of 5 Hz Sine Wave")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.xlim(0, 100)

#FFT of 50 Hz
plt.subplot(2, 1, 2)
plt.stem(freqs_50hz, fft_50hz, basefmt=" ")
plt.title("FFT of 50 Hz Sine Wave")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.xlim(0, 100)
```

```python
# E. Plotting normal random numbers
import numpy as np
import matplotlib.pyplot as plt

mean = 0
sd = 1
samples  = 1000

 

gd1 = np.random.normal(mean, sd, samples)
gd2 = np.random.normal(mean, sd, samples)
gd3 = np.random.normal(mean, sd, samples)


plt.plot(np.sort(gd1), np.arange(samples) )
plt.plot(np.sort(gd2), np.arange(samples) )
plt.plot(np.sort(gd1+gd2), np.arange(samples))
plt.plot(np.sort(gd1+gd2+gd3), np.arange(samples))
plt.xlabel("Values")
plt.ylabel("samples")
```
